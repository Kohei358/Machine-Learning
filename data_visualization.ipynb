# Matplotlib Core Visualization Methods

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.1, 100)
y2 = np.cos(x) + np.random.normal(0, 0.1, 100)

fig, ax = plt.subplots(figsize=(14, 10), constrained_layout=True)
# fig.suptitle('Matplotlib Core Visualization Methods', fontsize=16, fontweight='bold')

# --- Plot 1: Line & Scatter Combo ---
ax.plot(x, np.sin(x), color='gray', linestyle='--', label='Trend')
ax.scatter(x, y1, c=y1, cmap='viridis', edgecolors='black', alpha=0.8, label='Data')
ax.set_title('Line & Scatter Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend()
ax.grid(True, linestyle=':', alpha=0.6)
```

```python
ax.clear()

# --- Plot 2: Horizontal Bar Chart ---
fig, ax = plt.subplots(figsize=(14, 10), constrained_layout=True)
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [23, 17, 35, 29]
bars = ax.barh(categories, values, color='skyblue', edgecolor='navy')
ax.set_title('Horizontal Bar Chart')
ax.set_xlabel('Quantity')
ax.bar_label(bars, padding=3)

plt.show()
```

```python
# --- Plot 3: Histogram & Boxplot Combo ---
ax.clear()
fig, ax = plt.subplots(figsize=(14, 10), constrained_layout=True)
ax.hist(y1, bins=15, alpha=0.5, color='green', edgecolor='black', label='Hist Y1')
ax.hist(y2, bins=15, alpha=0.5, color='orange', edgecolor='black', label='Hist Y2')
ax.set_title('Histogram Distribution')
ax.legend()

plt.show()
```

```python
# --- Plot 4: Heatmap / Correlation Matrix ---
ax.clear()
fig, ax = plt.subplots(figsize=(14, 10), constrained_layout=True)
matrix_data = np.random.rand(5, 5)
im = ax.imshow(matrix_data, cmap='coolwarm', aspect='auto')
ax.set_title('Heatmap Matrix')
# Add a colorbar specifically linked to this axis
fig.colorbar(im, ax=ax)

print("matrix_data", matrix_data)
# plt.savefig('matplotlib_gallery.png', dpi=300) # Save the figure for your Git Repository README
plt.show()
```
